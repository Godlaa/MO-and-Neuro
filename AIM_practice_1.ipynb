```python
import pandas

data = pandas.read_csv('titanic.csv')

data
```

```python
female = data[data['Sex'] == 'female'].dropna()
male = data[data['Sex'] == 'male'].dropna()

print(female)
print(male)
```

```python
import matplotlib.pyplot as plt

data['Age'].plot(kind='hist', bins=50, alpha=0.7)
plt.xlabel('Возраст')
plt.ylabel('Количество людей')
```

```python
print(male['Age'].min())
print(male['Age'].max())
print(male['Age'].mean())
print(male['Age'].mode()[0])
print(male['Age'].median())
```

```python
print(female['Age'].min())
print(female['Age'].max())
print(female['Age'].mean())
print(female['Age'].mode()[0])
print(female['Age'].median())
```

```python
female['Age'].plot(kind='line', label='Мужчины', linestyle='-')
male['Age'].plot(kind='line', label='Женщины', linestyle='-')

plt.title('line')
plt.ylabel('Age')
plt.legend()
plt.show()
```

```python
male['Age'].plot(kind='hist', alpha=0.4, label='Мужчины', bins=10)
female['Age'].plot(kind='hist', alpha=0.4, label='Женщины', bins=10)

plt.title('hist')
plt.xlabel('Возраст')
plt.ylabel('Частота')
plt.legend()
plt.show()
```

```python
plt.scatter(male.index, male['Age'], label='Мужчины', alpha=0.7)
plt.scatter(female.index, female['Age'], label='Женщины', alpha=0.7)

plt.title('scatter')
plt.ylabel('Возраст')
plt.legend()
plt.show()
```
